```python
import numpy as np
import matplotlib.pyplot as plt
# import plotly.express as px 
import math
import pandas as pd
from random import *

num = 290
```

구간 분할

```python
def bound(temp):
    if temp == 25.1:
        b = 1
    elif temp == -3.9:
        b = 290
    elif temp > 25.1 or temp < -3.9:
        b = 999
    elif temp >= 0:
        b = (251 - int(temp * 10))
    elif temp <= 0:
        b = (251 + math.ceil(abs(temp * 10)))
        
    return b
```

기온 => 구간

```python
f = open("data.txt", "r")

temp = []

while True:
    line = f.readline()
    if not line:
        break
    sp = line.split("\t")
    sp2 = []
    for x in sp:
        sp2.append(float(x))
    temp.append(sp2)
    
temp = np.copy(temp)
part = np.copy(temp)

for i, _ in enumerate(temp):
    for j, x in enumerate(temp[i]):
        part[i,j] = bound(x)
    
part
```

```python
result = [[0]*num for i in range(num)]

for j in range(len(part[0])):
    for i in range(len(part) - 1):
        if(int(part[i][j]) == 999 or int(part[i+1][j]) == 999 ):
            continue
        x = result[int(part[i][j]) - 1][int(part[i+1][j]) - 1]
        if(int(part[i][j]) - 1 <= int(part[i+1][j]) - 1):
            result[int(part[i][j]) - 1][int(part[i + 1][j]) - 1] = x + 1
            


temp = 0
total = []

for i in range(len(result)):
    temp = 0.0
    for j in range(len(result[0])):
        temp += result[i][j]
    total.append(temp)
    

for i, _ in enumerate(result):
    for j, data in enumerate(result[i]):
            temp = data / total[i]
            result[i][j] = temp

mc = np.copy(result)

for i, _ in enumerate(mc):
    temp = 0
    for j, data in enumerate(mc[i]):
        temp += data
    print(temp)

mc

#for i, data in enumerate(total):
    #print(str(round(25.0 - 0.1 * i, 2)) + ", " + str(data))
        
```

```python
def bound_inv(state):
    x2 = round((252 - state) / 10.0, 2)
    x1 = round(x2 - 0.1, 2)
        #print('state: ' +  str(state_2) + ' x1: ' + str(x1) + ' x2: '  + str(x2))
    
    yield x1
    yield x2
```

```python
mc_cumulative = np.copy(mc_ot)

for i in range(len(mc_cumulative)):
    for j in range(len(mc_cumulative[0])):
        for k in range(1, len(mc_cumulative[0, 0])):
            mc_cumulative[i, j, k] += mc_cumulative[i, j, k - 1]
```

```python
state = {'state 1': [], 'state 2': [], 'temperature_simulated': []}
for t in df['temperature']:
    state_1 = bound(t)
    x = random()
    state_2 = 0
    if(state_1 == 999):
        state['state 1'].append(state_1)
        state['state 2'].append(state_1)
        state['temperature_simulated'].append(state_1)
        continue
    while(x > mc_cumulative[0, state_1, state_2]):
        state_2 += 1
    state['state 1'].append(state_1)
    state['state 2'].append(state_2)
    
    boundary = bound_inv(state_2)
    
    x1 = next(boundary)
    x2 = next(boundary)

    temp_sim = uniform(x1, x2)
    state['temperature_simulated'].append(temp_sim)

print(len(state['state 1']))
print(len(state['state 2']))
print(len(state['temperature_simulated']))
df_state = pd.DataFrame(state)
df_state
```
